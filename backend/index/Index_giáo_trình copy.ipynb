```python
import time
from google.api_core import exceptions
def summarize_2(prompt, text, max_retries=3):
    input_text = prompt + text
    print(f"Độ dài đầu vào: {len(input_text)} ký tự")
    for attempt in range(max_retries):
        try:
            response = model.generate_content(input_text)
            return response.text
        except exceptions.ResourceExhausted as e:
            print(f"Lỗi ResourceExhausted, thử lại lần {attempt + 1}/{max_retries}: {e}")
            time.sleep(30)  # Đợi 60 giây theo gợi ý từ log
    raise Exception("Hết số lần thử lại, không thể tạo nội dung")
```

```python
def split_text(text):
  start = text.find('[')
  end = text.rfind(']')
  return text[start:end+1]
```

```python
import pdfplumber

path = r"H:\DOCUMENT\chatbot\Data\Bỏ\Danh Sách Các Nguyên Hiệu Trưởng.pdf"
with pdfplumber.open(path) as pdf:
    full_text = "\n".join([page.extract_text() for page in pdf.pages[0:]])
print(full_text)
```

```python
import re

# Định nghĩa pattern
heading_pattern = r'^\d+\.\d+(?:\.\d+)?\.?\s*'

# Tách dữ liệu thành các đoạn dựa trên pattern
segments = re.split(heading_pattern, full_text, flags=re.MULTILINE)

headers = []
for i, segment in enumerate(segments):
    headers.append(segment.splitlines()[0].strip())

# Loại bỏ phần tử rỗng (nếu có) và kết hợp lại tiêu đề với nội dung
result = []
for i in range(len(segments)):
    if segments[i].strip():  # Kiểm tra nếu đoạn không rỗng
        # Nếu là đoạn đầu tiên hoặc có tiêu đề trước đó
        if i > 0 and re.match(heading_pattern, full_text.splitlines()[sum(len(s.splitlines()) for s in segments[:i])]):
            # Lấy tiêu đề từ dòng trước đó
            lines = full_text.splitlines()
            header_index = sum(len(s.splitlines()) for s in segments[:i])
            header = lines[header_index].strip()
            result.append(f"{header} {segments[i].strip()}")
        else:
            result.append(segments[i].strip())

# In kết quả
for i, segment in enumerate(result, 1):
    print(f"Đoạn {i}:\n{segment}\n{'-'*50}")
```
